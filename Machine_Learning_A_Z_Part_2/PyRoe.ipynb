# Multiple Linear Regression

## Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import statsmodels.api as sm
```

## Dataset

```python
dataset = pd.read_csv('roedata.csv')
X = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

## Encoding Categorical Data

## Spliting Dataset

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training model Linear

```python
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
print(regressor.intercept_)
print(regressor.coef_)
print(regressor.score(X_test, y_test))
print(regressor.score(X_train, y_train))
```

# Predicting Results

```python
y_pred = regressor.predict(X_test)
np.set_printoptions(precision=2)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1))
```

# Vis Training

```python
plt.scatter(y_train, y_train, color = 'red')
plt.scatter( y_test, y_test, color = 'yellow')
plt.plot(regressor.predict(X_train), regressor.predict(X_train), color = 'blue')
# write the formula of the regression line on the graph
plt.title('ROE vs Model')
plt.xlabel('ROE')
plt.ylabel('Model')
plt.show()
```

# Training Modelo Polynomial

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X_train)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y_train)
```

# Visualizing Chart

```python
plt.scatter(y, y, color = 'red')
plt.plot(lin_reg_2.predict(poly_reg.fit_transform(X)), lin_reg_2.predict(poly_reg.fit_transform(X)), color = 'blue')
plt.plot(regressor.predict(X), regressor.predict(X), color = 'blue')
plt.title('ROE vs Model')
plt.xlabel('ROE')
plt.ylabel('Model')
plt.show()
```
